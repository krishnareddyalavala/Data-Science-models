```python
qa_pipeline = pipeline("question-answering", model="bert-large-uncased-whole-word-masking-finetuned-squad")
```

```python
def extract_text_from_pdf(pdf_file):
    text = ""
    with fitz.open(stream=pdf_file.read(), filetype="pdf") as doc:
        for page in doc:
            text += page.get_text()
    return text
```

```python
def answer_from_pdf(pdf_file, question):
    context = extract_text_from_pdf(pdf_file)
    if not context.strip():
        return "Could not extract any text from the PDF."
    result = qa_pipeline(question=question, context=context[:4000])  # limit context for performance
    return f"Answer: {result['answer']} (Score: {round(result['score']*100, 2)}%)"
```

```python
!pip install gradio transformers PyMuPDF -q

import gradio as gr
import fitz
from transformers import pipeline
from io import BytesIO


qa_pipeline = pipeline("question-answering", model="bert-large-uncased-whole-word-masking-finetuned-squad")


def extract_text_from_pdf(uploaded_file):
    text = ""
    with fitz.open(stream=BytesIO(uploaded_file.read()), filetype="pdf") as doc:
        for page in doc:
            text += page.get_text()
    return text.strip()


def qa_from_pdf(pdf_file, question):
    context = extract_text_from_pdf(pdf_file)
    if not context:
        return " Could not extract text from the PDF. Please upload a text-based PDF (not scanned image)."
    # Limit context to first 4000 characters to prevent model overload
    context = context[:4000]
    result = qa_pipeline(question=question, context=context)
    return f" Answer: {result['answer']} (Confidence: {round(result['score']*100, 2)}%)"


interface = gr.Interface(
    fn=qa_from_pdf,
    inputs=[
        gr.File(label="Upload PDF Document", file_types=[".pdf"]),
        gr.Textbox(label="Ask a Question", placeholder="e.g., What is the purpose of the document?")
    ],
    outputs="text",
    title=" BERT QA from PDF",
    description="Upload a text-based PDF and ask any question based on its content. Uses BERT fine-tuned on SQuAD."
)

interface.launch()
```
